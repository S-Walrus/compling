```python
import pickle
from collections import Counter
from pymystem3 import Mystem
from nltk.corpus import stopwords as sw
stopwords = sw.words('russian')
```

```python
with open('age_dump', 'rb') as f:
    d = pickle.load(f)
```

```python
stem = Mystem()
s = {key: Counter(stem.lemmatize(d[key])) for key in d.keys()}
```

```python
for key in s.keys():
    for word in stopwords:
        del s[key][word]
```

```python
r = {key: list(sorted(s[key].keys(), key=lambda x: s[key][x], reverse=True)) for key in s.keys()}
```

```python
r = {key: [word for word in r[key] if word.isalpha()] for key in r.keys()}
```
