```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('language.csv') as f:
    plain = f.read()
    print(',SOV,' in plain)
```

```python
data = pd.read_csv('language.csv', sep=',')
```

```python
col = '81A Order of Subject, Object and Verb'
```

```python
a = [item.split()[1] for item in data[col] if len(str(item).split()) >= 2]
```

```python
x = {item : a.count(item) for item in set(a)}
```

```python
fig, ax = plt.subplots()
fig.dpi = 100
ax.pie(x.values(), labels=x.keys())
ax.axis('equal')
plt.show()
```

```python
for item in x.keys():
    print(item + ' occurs ' + str(x[item]) + ' times')
```
