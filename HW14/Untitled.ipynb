```python
with open('pelevin.txt', encoding='CP1251') as f:
    pel_text = f.read()

with open('ivanov.txt', encoding='KOI8-R') as f:
    iva_text = f.read()
```

```python
import re
iva_text = '\n\n'.join(re.split(r'<ul>.*?</ul>', iva_text))
```

```python
from pymystem3 import Mystem
```

```python
stem = Mystem()
```

```python
type(iva_text)
```

```python
iva_an = stem.analyze(iva_text)
pel_an = stem.analyze(pel_text)
```

```python
iva_an = [item for item in iva_an if 'analysis' in item.keys()]
pel_an = [item for item in pel_an if 'analysis' in item.keys()]
```

```python
iva_an = [item for item in iva_an if len(item['analysis']) != 0]
pel_an = [item for item in pel_an if len(item['analysis']) != 0]
```

```python
pel_adv = len([item for item in pel_an if 'ADV' in item['analysis'][0]['gr']]) / len(pel_an)
iva_adv = len([item for item in iva_an if 'ADV' in item['analysis'][0]['gr']]) / len(iva_an)
```

```python
print('ADV in Pel: ' + str(pel_adv))
print('ADV in Iva: ' + str(iva_adv))
```

```python
pel_s = len([item for item in pel_an if 'S' in item['analysis'][0]['gr']]) / len(pel_an)
iva_s = len([item for item in iva_an if 'S' in item['analysis'][0]['gr']]) / len(iva_an)
```

```python
print('S in Pel: ' + str(pel_s))
print('S in Iva: ' + str(iva_s))
```

```python
pel_spro = len([item for item in pel_an if 'SPRO' in item['analysis'][0]['gr']]) / len(pel_an)
iva_spro = len([item for item in iva_an if 'SPRO' in item['analysis'][0]['gr']]) / len(iva_an)
```

```python
print('SPRO in Pel: ' + str(pel_spro))
print('SPRO in Iva: ' + str(iva_spro))
```

```python
pel_v = len([item for item in pel_an if 'V' in item['analysis'][0]['gr']]) / len(pel_an)
iva_v = len([item for item in iva_an if 'V' in item['analysis'][0]['gr']]) / len(iva_an)
```

```python
print('V in Pel: ' + str(pel_v))
print('V in Iva: ' + str(iva_v))
```

```python
pel_noun = len([item for item in pel_an if 'N' in item['analysis'][0]['gr']]) / len(pel_an)
iva_noun = len([item for item in iva_an if 'N' in item['analysis'][0]['gr']]) / len(iva_an)
```
