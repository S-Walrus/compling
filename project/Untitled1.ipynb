```python
import pickle
from random import choice
```

```python
with open('comment_l', 'rb') as f:
    com_l = pickle.load(f)
```

```python
com_l.sort(key=lambda x: x[1], reverse=True)
```

```python
com_l
```

```python
rimus = [item[0] for item in com_l if '—Ä–∏–º–ª—è–Ω' in item[0].lower()]
```

```python
print('\n\n\n'.join(rimus))
```

```python
for i, item in enumerate(rimus):
    f = item.lower().find('—Ä–∏–º–ª—è–Ω')
    rimus[i] = item[:f] + item[f:].split(' ', 1)[0].upper() + ' ' + (item[f:]+' ').split(' ', 1)[1]
```

```python
print('\n\n'.join(rimus))
```

```python
data = [choice(rimus) for i in range(200)]
```

```python
print('\n\n############\n\n'.join(data))
```
